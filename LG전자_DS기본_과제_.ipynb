```python
# 필요한 모듈을 임포트 합니다.
# 2) datetime 모듈을 datetime 클래스를 임포트하세요.
import datetime
from dateutil.relativedelta import relativedelta
```

```python
# 3) datetime 클래스를 이용해 현재 시간을 now 변수에 저장하세요.
now = datetime.datetime.now()

before = now - relativedelta(years=10)

# 현재 시간을 기준으로 시작 시간과 종료 시간을 문자열로 만들기
now_day = now.strftime("%Y-%m-%d")
befor_day = before.strftime("%Y-%m-%d")
print(f"end  : {now_day}")
print(f"start: {befor_day}")

# 4) get_data_yahoo 메서드를 이용해서 LG 전자의 주가데이터를 불러옵니다.
lg_stock = yf.download("066570.KS", start=befor_day, end=now_day)

# 주가 데이터가 잘 불러와졌는지 확인합니다.
print(lg_stock)

lg_stock.columns

```

```python
# 5) 데이터 프레임의 데이터를 확인하기 위해 일부 데이터를 출력해보세요.


#lg_stock.reset_index(level=[0])

colume_names = lg_stock.columns.get_level_values(0)
lg_stock.columns = colume_names

#lg_stock.columns
lg_stock.head()
```

```python
# 8) 현재 인덱스를 다시 데이터 열로 만드세요.
lg_stock.reset_index()
```

```python
# 9) 데이터 프레임의 컬럼명을 변경 해보세요.
"""
Data : 일자
Open : 시가
High : 고가
Low : 저가
Close : 종가
"""
```

```python
lg_stock = lg_stock.rename(columns={'Date':'일자','High':'고가','Low':'저가','Open' : '시가','Close' : '종가' })
lg_stock.reset_index()
```

```python
# 변경이 잘 되었는지 확인해봅니다.
lg_stock.head()
```

```python
# 10) 일자 컬럼을 다시 인덱스로 만드세요.
lg_stock.reset_index('Date', inplace=True)
#lg_stock.reset_index('Date')
```

```python
# 잘 변경되었는지 확인해봅시다.
lg_stock.tail()
```

```python
# 새로운 컬럼을 만들어서 데이터를 추가한다!
# 11) 5일씩 묶어서 5일이평, 10일씩 묶어서 10일이평 ...컬럼을 만들어 추가해 보세요.
# 5일, 10일, 15일, 20일, 30일, 60일
# * 이평이란 해당 일자 만큼의 종가 가격을 합쳐 평균을 낸 것을 의미합니다.
new_lg_stock = lg_stock

ma5 = lg_stock['Adj Close'].rolling(window=5).mean()
ma10 = lg_stock['Adj Close'].rolling(window=10).mean()
ma15 = lg_stock['Adj Close'].rolling(window=15).mean()
ma20 = lg_stock['Adj Close'].rolling(window=20).mean()
ma30 = lg_stock['Adj Close'].rolling(window=30).mean()
ma60 = lg_stock['Adj Close'].rolling(window=60).mean()

new_lg_stock.insert(len(lg_stock.columns), "MA5", ma5)
new_lg_stock.insert(len(lg_stock.columns), "MA10", ma10)
new_lg_stock.insert(len(lg_stock.columns), "MA15", ma15)
new_lg_stock.insert(len(lg_stock.columns), "MA20", ma20)
new_lg_stock.insert(len(lg_stock.columns), "MA30", ma30)
new_lg_stock.insert(len(lg_stock.columns), "MA60", ma60)

new_lg_stock.head(60)
```

```python
# 생성이 잘 되었나 확인해봅시다.
lg_stock.head()
```

```python
# 오늘 것들끼리 비교, 어제 것들끼리 비교
# 컬럼의 데이터끼리 비교를 수행할 수 있습니다.
#lg_stock['5일이평'] > lg_stock['30일이평']
import pandas as pd
df = pd.DataFrame({'5일이평': new_lg_stock['MA5'],
                   '30일이평': new_lg_stock['MA10']})

df['결과'] = new_lg_stock['MA5'] > new_lg_stock['MA10']
df
```

```python
# 12) 5일 이평과 30일 이평을 한 로우씩 아래로 내려 어제5일이평, 어제30일이평 이라는 컬럼을 추가해보세요.
new_lg_stock['MA5-1'] = new_lg_stock['MA5'].shift(1)
new_lg_stock['MA30-1'] = new_lg_stock['MA30'].shift(1)
new_lg_stock.head(31)
```

```python
# 13) 5일이평, 어제5일이평, 30일이평, 어제30일이평 4개 컬럼을 이용해서 골든 데드 크로스가 발생하는지 여부를 확인해보세요.
# 골든 (어제는 단기 이평이 장기 이평보다 낮았는데, 오늘은 단기 이평이 장기 이평보다 높다.)
new_lg_stock['is_golden'] = (new_lg_stock['MA5'] > new_lg_stock['MA30']) > (new_lg_stock['MA5-1'] < new_lg_stock['MA30-1'])
```

```python
# 데드 (어제는 단기 이평이 장기 이평보다 높았는데, 오늘은 단기 이평기 장기 이평보다 낮다.)
new_lg_stock['is_dead'] = (new_lg_stock['MA5'] > new_lg_stock['MA30']) > (new_lg_stock['MA5-1'] > new_lg_stock['MA30-1'])
```

```python
# 데이터가 잘 생성되었는지 확인해봅시다.
new_lg_stock.head()
```

```python
# 14) 필요 없는 Adj Close, Volume, 20일이평, 60일이평 컬럼을 삭제해보세요.
type(new_lg_stock)
tmp_lg_stock = new_lg_stock.drop(['Volume', 'Adj Close','MA20','MA60'], axis=1)
tmp_lg_stock.head()
```

```python
# 15) '종가','5일이평','30일이평','어제5일이평', '어제30일이평','is_golden', 'is_dead' 만 남겨놓고 나머지 컬럼을 다 삭제하세요.
tmp2_lg_stock = tmp_lg_stock.drop(['Date','고가','저가','시가','MA10','MA15'], axis=1)
tmp2_lg_stock
```

```python
# 16) 시각화 모듈은 matplotlib을 import 하세요.
import matplotlib.pyplot as plt
```

```python
# 17) 아래 그래프 그림을 참고하여 그래프를 그려보세요.
# 5일이평, 30일이평 꺾은선 그래프
# Golden 크로스 발생 지점은 gold 컬러로 삼각형 마커
# Dead 크로스 발생 지점은 black 컬러로 역삼각형 마크
# 범례는 best 위치에 표기
# 그리드 표시
# 타이틀 : LG Stock Graph
# x축 레이블 : Date
# y축 레이블 : Price
# 타이트 레이아웃 사용하기

plt.plot(tmp2_lg_stock.index, tmp2_lg_stock['MA5'],label='MA5')
plt.plot(tmp2_lg_stock.index, tmp2_lg_stock['MA30'],label='MA30')

for i in range(len(tmp2_lg_stock)):
  if tmp2_lg_stock['is_golden'][i] == True:
    plt.scatter(tmp2_lg_stock.index, tmp2_lg_stock['MA5'],color='gold', marker='^')
  if tmp2_lg_stock['is_dead'][i] == True:
    plt.scatter(tmp2_lg_stock.index, tmp2_lg_stock['MA5'],color='black', marker='v')

plt.xlabel('Date')
plt.ylabel('price')
plt.title('LG Stock Graph')
plt.legend(loc='upper right')
```
